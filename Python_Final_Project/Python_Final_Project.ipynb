* –û—Å—Ç–∞–≤—å—Ç–µ –≤ —Ç–∞–±–ª–∏—Ü–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é —Ç–æ–ª—å–∫–æ –¥–ª—è —Ç–µ—Ö –ª–µ—Ç, –¥–ª—è –∫–æ—Ç–æ—Ä—ã—Ö –µ—Å—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –±–æ–ª–µ–µ —á–µ–º 50 —Ä–∞—É–Ω–¥–∞—Ö —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è.

* –ù–∞ –æ—Å–Ω–æ–≤–µ –ø–æ–ª—É—á–∏–≤—à–µ–π—Å—è —Ç–∞–±–ª–∏—Ü—ã –ø–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫, –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç –æ—Ç—Ä–∞–∂–∞—Ç—å –¥–∏–Ω–∞–º–∏–∫—É —Ç–∏–ø–∏—á–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ —Å—Ä–µ–¥—Å—Ç–≤, –∫–æ—Ç–æ—Ä—ã–µ —Å—Ç–∞—Ä—Ç–∞–ø—ã –ø–æ–ª—É—á–∞–ª–∏ –≤ —Ä–∞–º–∫–∞—Ö –æ–¥–Ω–æ–≥–æ —Ä–∞—É–Ω–¥–∞ —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è.

–ù–∞ –æ—Å–Ω–æ–≤–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –æ—Ç–≤–µ—Ç—å—Ç–µ –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã:

* –í –∫–∞–∫–æ–º –≥–æ–¥—É —Ç–∏–ø–∏—á–Ω—ã–π —Ä–∞–∑–º–µ—Ä —Å–æ–±—Ä–∞–Ω–Ω—ã—Ö –≤ —Ä–∞–º–∫–∞—Ö –æ–¥–Ω–æ–≥–æ —Ä–∞—É–Ω–¥–∞ —Å—Ä–µ–¥—Å—Ç–≤ –±—ã–ª –º–∞–∫—Å–∏–º–∞–ª–µ–Ω?

* –ö–∞–∫–∞—è —Ç–µ–Ω–¥–µ–Ω—Ü–∏—è –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Ä–∞—É–Ω–¥–æ–≤ –∏ –≤—ã–¥–µ–ª—è–µ–º—ã—Ö –≤ —Ä–∞–º–∫–∞—Ö –∫–∞–∂–¥–æ–≥–æ —Ä–∞—É–Ω–¥–∞ —Å—Ä–µ–¥—Å—Ç–≤ –Ω–∞–±–ª—é–¥–∞–ª–∞—Å—å –≤ 2013 –≥–æ–¥—É?

–Ω–∞–∏–±–æ–ª—å—à–∏–º–∏ —Ç–∏–ø–∏—á–Ω—ã–º–∏ —Ü–µ–Ω–∞–º–∏;
–∏ –Ω–∞–∏–±–æ–ª—å—à–∏–º —Ä–∞–∑–±—Ä–æ—Å–æ–º —Ü–µ–Ω –∑–∞ —Å—Ç–∞—Ä—Ç–∞–ø.
–û–±—ä—è—Å–Ω–∏—Ç–µ, –ø–æ—á–µ–º—É —Ä–µ—à–∏–ª–∏ —Å–æ—Å—Ç–∞–≤–∏—Ç—å —Ç–æ–ø –∏–º–µ–Ω–Ω–æ –∏–∑ —Ç–∞–∫–æ–≥–æ —á–∏—Å–ª–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏–π –∏ –ø–æ—á–µ–º—É —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–ª–∏ –∏–º–µ–Ω–Ω–æ —Ç–∞–∫.

```python
df_filtered = df_final_merged[df_final_merged['funding_total'] > 0]

q1 = df_filtered['funding_total'].quantile(0.25)
q2 = df_filtered['funding_total'].quantile(0.50)
q3 = df_filtered['funding_total'].quantile(0.75)

def categorize_funding(value):
    if value <= q1:
        return '–ù–∏–∑–∫–æ–µ'
    elif value <= q2:
        return '–°—Ä–µ–¥–Ω–µ–µ'
    elif value <= q3:
        return '–í—ã—Å–æ–∫–æ–µ'
    else:
        return '–û—á–µ–Ω—å –≤—ã—Å–æ–∫–æ–µ'
        
df_filtered = df_filtered.copy()
df_filtered['funding_category'] = df_filtered['funding_total'].apply(categorize_funding)

df_category_stats = df_filtered.groupby('funding_category')['funding_total'].agg(
    median_price='median',
    iqr=lambda x: x.quantile(0.75) - x.quantile(0.25) #—Ä–µ—à–∏–ª, —á—Ç–æ –∑–¥–µ—Å—å –±—É–¥–µ—Ç —É–¥–æ–±–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ª—è–º–±–¥–∞ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è —Ä–∞–∑–º–∞—Ö–∞ –≤–Ω—É—Ç—Ä–∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
).reset_index()

df_sorted = df_category_stats.sort_values(by='median_price', ascending=False)
df_sorted.head()
```
