```python
import matplotlib.pyplot as plt
import pandas as pd

tmp = df_merged[df_merged.seats != -1].copy()  # –°–æ–∑–¥–∞–µ–º –∫–æ–ø–∏—é –¥–∞—Ç–∞—Å–µ—Ç–∞, —á—Ç–æ –±—ã –Ω–µ —Ä–∞–±–æ—Ç–∞—Ç—å —Å –∏—Å—Ö–æ–¥–Ω—ã–º

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –º–µ–¥–∏–∞–Ω—ã –¥–ª—è –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
medians = tmp.groupby('category')['seats'].median().sort_values(ascending=False)

# –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø–æ –º–µ–¥–∏–∞–Ω–µ
sorted_categories = medians.index
tmp['category'] = pd.Categorical(tmp['category'], categories=sorted_categories, ordered=True)

tmp.boxplot(column='seats', 
             by='category', 
             grid=True,       # –í—ã–≤–æ–¥–∏–º —Å–µ—Ç–∫—É –Ω–∞ –¥–∏–∞–≥—Ä–∞–º–º—É
             showmeans=True,  # –ó–µ–ª–µ–Ω—ã–π —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∏ –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ  
             figsize=(10, 6))

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –∑–∞–≥–æ–ª–æ–≤–∫–∞ –∏ –º–µ—Ç–æ–∫ –æ—Å–µ–π
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏–∏')
plt.xlabel('–ù–∞–∑–≤–∞–Ω–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –∑–∞–≤–µ–¥–µ–Ω–∏—è')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç')

plt.xticks(rotation = 30)  #  –ü–æ–≤–æ—Ä–∞—á–∏–≤–∞–µ–º –º–µ—Ç–∫–∏ –Ω–∞ –æ—Å–∏ –•, —á—Ç–æ –±—ã –æ–Ω–∏ –Ω–µ —Å–ª–∏–≤–∞–ª–∏—Å—å 
plt.ylim(0,330)            # –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º –≤—ã–≤–æ–¥ –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –±–æ–∫—Å–ø–ª–æ—Ç—ã –±—ã–ª–∏ –∫—Ä—É–ø–Ω—ã–º–∏, 
                           # –µ—Å–ª–∏ —É–±—Ä–∞—Ç—å, —Ç–æ –±—É–¥—É—Ç –≤–∏–¥–Ω—ã –≤—Å–µ –≤—ã–±—Ä–æ—Å—ã
plt.suptitle('')           # –£–¥–∞–ª–µ–Ω–∏–µ –º–µ—Ç–∫–∏ "seats" —Å –æ—Å–∏ X
plt.show()                 # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –¥–∏–∞–≥—Ä–∞–º–º—ã
```

### –ó–∞–¥–∞—á–∞ 6

–ò–∑—É—á–∏—Ç–µ, —Å –∫–∞–∫–∏–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç —Å–∞–º—É—é —Å–∏–ª—å–Ω—É—é –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é —Ä–µ–π—Ç–∏–Ω–≥–∏ –∑–∞–≤–µ–¥–µ–Ω–∏–π? –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –∏ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–π—Ç–µ –º–∞—Ç—Ä–∏—Ü—É –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —Ä–µ–π—Ç–∏–Ω–≥–∞ –∑–∞–≤–µ–¥–µ–Ω–∏—è —Å —Ä–∞–∑–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏: –µ–≥–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è, –ø–æ–ª–æ–∂–µ–Ω–∏–µ (–∞–¥–º–∏–Ω–∏—Å—Ç—Ä–∞—Ç–∏–≤–Ω—ã–π —Ä–∞–π–æ–Ω –ú–æ—Å–∫–≤—ã), —Å—Ç–∞—Ç—É—Å —Å–µ—Ç–µ–≤–æ–≥–æ –∑–∞–≤–µ–¥–µ–Ω–∏—è, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–µ—Å—Ç, —Ü–µ–Ω–æ–≤–∞—è –∫–∞—Ç–µ–≥–æ—Ä–∏—è –∏ –ø—Ä–∏–∑–Ω–∞–∫, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –∑–∞–≤–µ–¥–µ–Ω–∏—è –∫—Ä—É–≥–ª–æ—Å—É—Ç–æ—á–Ω—ã–º. –í—ã–±–µ—Ä–∏—Ç–µ —Å–∞–º—É—é —Å–∏–ª—å–Ω—É—é —Å–≤—è–∑—å –∏ –ø—Ä–æ–≤–µ—Ä—å—Ç–µ –µ—ë.

```python
df_for_phik = df_merged[['category', 'district', 'chain', 'price', 'is_24_7', 'rating']]
correlation_matrix = df_for_phik.phik_matrix(interval_cols=['rating'])
rating_corr = correlation_matrix.loc[correlation_matrix.index != 'rating'][['rating']].sort_values(by='rating', ascending=False)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ —Å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–º phi_k –¥–ª—è –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π rating')
```

```python
print(rating_corr)
```

```python
df_filtered['middle_avg_bill'].plot(
    kind = 'box',
    vert = False
)
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞')
plt.xlabel('–°—Ä–µ–¥–Ω–∏–π —á–µ–∫')
plt.show()
```

```python
avg_bill_by_district = df_filtered.groupby('district')['middle_avg_bill'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 8))
avg_bill_by_district.plot(kind='barh', color='skyblue')
plt.title('–°—Ä–µ–¥–Ω–∏–π —á–µ–∫ –ø–æ —Ä–∞–π–æ–Ω–∞–º –ú–æ—Å–∫–≤—ã')
plt.xlabel('–°—Ä–µ–¥–Ω–∏–π —á–µ–∫')
plt.ylabel('–†–∞–π–æ–Ω—ã')
plt.gca().invert_yaxis()
plt.show()
```

```python
avg_bill_by_district = df_filtered.groupby('district')['middle_avg_bill'].mean().sort_values(ascending=False)
mean_bill_by_district = df_filtered.groupby('district')['middle_avg_bill'].median().sort_values(ascending=False)
pivot_table = pd.DataFrame({
    '–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ': avg_bill_by_district,
    '–ú–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ': mean_bill_by_district
}).reset_index()
print(pivot_table)
```
